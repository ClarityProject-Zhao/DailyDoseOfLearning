# EDA of Word Happiness Report 2020

The main goal of this project is to get familiar with seaborn package. In this analysis, I will focus on EDA using a popular data set from Kaggle community - the World Happiness Report.

1. Data Import

```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import importlib

folder = os.getcwd()
data_2020 = pd.read_csv(os.path.join(folder, r'data/2020.csv'))
```

```python
data_2020.head(5)
```

2. Data cleaning

There is not much cleaning needed as the data in this file is well organized. I add another region column in the dataframe with shorter text for plotting later.

```python
dict_region = {'Western Europe': 'W.Eur',
               'North America and ANZ': 'NA,ANZ',
               'Middle East and North Africa': "Mid.East, N.Africa",
               'Latin America and Caribbean':'SA',
               'Central and Eastern Europe': 'Cen.E,EUR',
               'East Asia': 'E.Asia',
               'Southeast Asia': 'S.Asia',
               'Commonwealth of Independent States': 'Ind.States',
               'Sub-Saharan Africa': 'SS.Africa',
               'South Asia': 'S.Asia'
               }
data_2020['region']=data_2020['Regional indicator'].apply(lambda x: dict_region[x])
```

```python
data_2020['region']
```

3. Plotting

Let's start with bar plot

```python
plt.rcParams.update({'figure.autolayout': True})
ax_bar = sns.barplot(data=data_2020, x='region', y='Ladder score')  # axe level object
ax_bar.set_xticklabels(ax_bar.get_xticklabels(), rotation=40, horizontalalignment='right')
```

While the plot shows the average happiness score of different regions, it does not show how many countries are within each region and what the distribution looks like. We use countplot and distribution plot to get a better sense of these details

```python
plt.rcParams.update({'figure.autolayout': True})
ax_count = sns.countplot(data=data_2020, x='region')  # axe level object
ax_count.set_xticklabels(ax_count.get_xticklabels(), rotation=40, horizontalalignment='right')
```

```python
plt.rcParams.update({'figure.autolayout': True})
fig_violin = sns.catplot(data=data_2020, x='region', y='Ladder score', kind='violin')
fig_violin.set_xticklabels(fig_violin.axes[0, 0].get_xticklabels(), rotation=40, horizontalalignment='right')
```

Also a close up of histgram within each region group

```python
#displot by region
plt.rcParams.update({'figure.autolayout': True})
fig_dis=sns.displot(data=data_2020, x="Ladder score", col="region", legend=False)
fig_dis.set(xlabel=None)
```

Not a surprise in western Europe, NA and ANZ regions, happiness score skews largely to the right. The scores are more centrally distributed in other regions with higher std and varied means.

Now we include other fields that are part of the happiness score survey, to understand the relationship of happiness score it is potential drivers. I start with a pairwise distribution plot

```python
plt.rcParams.update({'figure.autolayout': True})
data_score=data_2020[['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
g = sns.PairGrid(data_score)
g.map_upper(sns.regplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, kde=True)
```

GDP, social support, healthy life expectancy and freedom to make life choices seem to correlate positively with happines score. Corruption on the other hand seems to correlate negatively. Relationship between happiness and generosity is somewhat ambiguous. Is this also the case within each region?

```python
g1=sns.lmplot(data=data_2020,x="Logged GDP per capita", y="Ladder score",col='region')
g1.set(xlabel=None)
g2=sns.lmplot(data=data_2020,x="Social support", y="Ladder score",col='region')
g2.set(xlabel=None)
g3=sns.lmplot(data=data_2020,x="Healthy life expectancy", y="Ladder score",col='region')
g3.set(xlabel=None)
g4=sns.lmplot(data=data_2020,x="Freedom to make life choices", y="Ladder score",col='region')
g4.set(xlabel=None)
g5=sns.lmplot(data=data_2020,x="Generosity", y="Ladder score",col='region')
g5.set(xlabel=None)
g6=sns.lmplot(data=data_2020,x="Perceptions of corruption", y="Ladder score",col='region')
g6.set(xlabel=None)
```

The problem of this breakdown is that often times countries within a region may have very different social structures hence not comparable. The other problem some region has a small data set of countries within, for example in NA/ANZ group. To get around this, I use k-mean clustering to re-group the data. First off though, I need to decide how many neighbors I want to use. I start with the elbow approach.

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics.pairwise import pairwise_distances_argmin

data_x=data_2020[['Logged GDP per capita', 'Social support', 'Healthy life expectancy',
                        'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
#need to decide n_clusters
#Elbow method
WSS={}
for i in range(2,30):
    k_means = KMeans(init='k-means++', n_clusters=i, n_init=10)
    k_means.fit(data_x)
    k_means_labels, k_means_distance = metrics.pairwise_distances_argmin_min(data_x, k_means.cluster_centers_)
    WSS[i]=np.sum(k_means_distance**2)
WSS_pd=pd.DataFrame(WSS.values(),index=WSS.keys(),columns=['WSS'])
WSS_pd.plot()
```

Unfortunately, the answer is not so clear. I turn to silhouette method instead.

```python
#no clean answer
from sklearn.metrics import silhouette_score
sil={}
for i in range(2,30):
    k_means = KMeans(init='k-means++', n_clusters=i, n_init=10)
    k_means.fit(data_x)
    k_means_labels, k_means_distance = metrics.pairwise_distances_argmin_min(data_x, k_means.cluster_centers_)
    sil[i] = silhouette_score(data_x, k_means_labels, metric='euclidean')
sil_pd=pd.DataFrame(sil.values(),index=sil.keys(),columns=['silhouette'])
sil_pd.plot()
```

Silhouette method suggests 2 is the optimal number of clusters we should use.

Not sure whether the result is impacted by the high dimensionality, I reduce the dimension using PCA and test the grouping again.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_x_pca = pca.fit_transform(data_x)

sil_pca = {}
for i in range(2, 30):
    k_means = KMeans(init='k-means++', n_clusters=i, n_init=10)
    k_means.fit(data_x_pca)
    k_means_labels, k_means_distance = metrics.pairwise_distances_argmin_min(data_x_pca, k_means.cluster_centers_)
    sil_pca[i] = silhouette_score(data_x_pca, k_means_labels, metric='euclidean')
sil_pca_pd = pd.DataFrame(sil_pca.values(), index=sil_pca.keys(), columns=['silhouette'])
sil_pca_pd.plot()
```

Two is still the winner! While this is not exactly what I expected, the number makes sense. In a sense, this is somewhat similar to how we perceive the world these days, i.e. categorizing the countries into developed and less developed. This also makes me think, maybe I should not try to group the countries at all - according to the 'Factfulness', we human tends to see the world as binary even with gradual changes and incremental differences. For the sake of practice, I will leave the group as 2 for now. Now let's visualize the two clusters using the lower demension data.

```python
k_means = KMeans(init='k-means++', n_clusters=2, n_init=10)
k_means.fit(data_x)
k_means_labels, k_means_distance = metrics.pairwise_distances_argmin_min(data_x, k_means.cluster_centers_)

data_2020.loc[:, 'KN_label'] = pd.Series(k_means_labels)
data_2020.loc[:, 'KN_distance'] = pd.Series(k_means_distance)
data_2020[['Country name', 'KN_label']]

# plot cluster in 2D surface
data_pca_results = pd.DataFrame(data_x_pca, columns=['x1', 'x2']).join(pd.DataFrame(k_means_labels, columns=['label']))
sns.jointplot(data=data_pca_results, x='x1', y='x2', hue='label').plot_joint(sns.kdeplot)
```

While the intuition tells that the two clusters could very well be the 'developed' and 'less developed' country clusters. Let's check if this is indeed the case.

```python
k_means = KMeans(init='k-means++', n_clusters=2, n_init=10)
k_means.fit(data_x)
k_means_labels, k_means_distance = metrics.pairwise_distances_argmin_min(data_x, k_means.cluster_centers_)

data_2020.loc[:,'KN_label']=pd.Series(k_means_labels)
data_2020.loc[:,'KN_distance']=pd.Series(k_means_distance)
data_2020[['Country name','KN_label']]
```

Seems to me that results are in line with intuition.

In addition, the clustering plot suggests that the classification is mainly driven by one variable, i.e. the first PCA component. Curious to see what comprises variable 1, I decide to further calculate the loadings.

```python
# loadings
loadings=pca.components_.T * np.sqrt(pca.explained_variance_)
loadings_pd=pd.DataFrame(loadings.T,columns=data_x.columns.tolist())
loadings_pd.plot(kind='bar')
```

While it is not a surprise GDP is one of the main drivers, it is interesting to see life expectancy is a even more important driver of the grouping. Given that we essentially group countries into 'developed' and 'less developed' regions, this suggests healthy life expectancy is one of the key divider between these two regions.

Now let's see whether our analysis will be different using two-group classification instead of geography classification.

Different from what we saw previously, the new breakdown suggests 'generosity' factor contributes to happiness in 'developed' countries quite meaningfully. In addition, while perception of corruption impacts happiness in a negative way in the 'developed' world, it does not have significant impact on happiness in less developed countries.

I further decide to take a look at the trend of hapiness score over years. First I need to combine inputs from different years. As data files are structured differently, some cleaning is needed.

```python
data={}
data_all=pd.DataFrame()
for year in range(2015,2021):
    data[year] = pd.read_csv(os.path.join(folder, r'data/',f'{str(year)}.csv'))
    data[year].rename(columns={'Country name': 'Country', 'Country or region': 'Country'}, inplace=True)
    data[year].rename(columns={'Ladder score': f'{str(year)}', 'Happiness Score': f'{str(year)}',
                               'Happiness.Score': f'{str(year)}','Score': f'{str(year)}'},
                      inplace=True)
    data_all=pd.merge(data_all,data[year][['Country',f'{str(year)}']],on='Country') if not data_all.empty else data[year][['Country',f'{str(year)}']]
data_combo=pd.merge(data_all,data_2020[['Country name','region','KN_label']],left_on='Country',right_on='Country name')
data_combo.drop(columns='Country name',inplace=True)
data_combo_long=pd.melt(data_combo, id_vars=['Country','region','KN_label'], value_vars=[f'{str(x)}' for x in range(2015,2021)])
data_combo_long.columns=['Country','Region','KN_label','Year','Score']
```

While we have mostly focused on using 2020 survey data. I want to include the previous years to see how the trend looks like.

```python
sns.relplot(data=data_combo_long,x='Year',y='Score',kind='line')
```

From a global point of view, people are happier over the past few years. Is that the case across region? I use both the 2-cluster groupping and the geography based groupping to break down the global score.

```python
sns.relplot(data=data_combo_long,x='Year',y='Score',col='KN_label',kind='line')
```

```python
g=sns.relplot(data=data_combo_long,x='Year',y='Score',col='Region',col_wrap=2,hue='Country',legend=False)
```

Interestingly, while overall global happiness is one the rise, the increase is mostly driven by 'less developed' countries. At regional level, countries in sub-saharan Africa and middle/east Europe experienced an gradual improvement in happiness while the trend is somewhat the opposite in NA/ANZ. We can also see what happens in each specific country as below.

```python
fig = plt.figure(figsize=(20, 40))
plt.tight_layout()
axes = fig.add_gridspec(5, 2)

for i, region_str in enumerate(dict_region.values()):
    ax=fig.add_subplot(axes[i // 2, i % 2])
    sns.pointplot(data=data_combo_long.loc[data_combo_long['Region'] == region_str, :], x='Year', y='Score',
                  col='Region', col_wrap=2, hue='Country', ax=ax)
    ax.legend(ncol=4,framealpha=0.3)
    ax.set_ylim(int(data_combo_long['Score'].min()),int(data_combo_long['Score'].max()+1))
```
